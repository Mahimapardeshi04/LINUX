   software that manages all of the hardware and software on a computer. It’s similar to Windows or macOS, but it’s free to use, highly customizable, and especially popular among developers, system administrators, and in server environments.

2. What is the difference between Hard Link & Soft Link?
   - **hard link**:-A Hard Link is essentially an exact copy of a file — not in the sense of copying its content, but rather it points to the same data blocks on the disk as the original file. In Linux, every file is actually just a link to something called an inode, which is a data structure that stores the file’s actual content and metadata.

   - When you create a hard link, you're creating a new filename that links to the same inode as the original file. That means the hard-linked file and the original file are indistinguishable — both are just names for the same underlying data.

  -> **softlink:**-A Soft Link, or Symbolic Link, is more like a shortcut or alias. It doesn’t point to the actual data, but rather to the filename of another file. You can easily recognize a symbolic link by the arrow notation (->) in the terminal.
  - They can link to directories

  - They can span across different file systems

  - They are easier to recognize and manage for symbolic referencing


3.  What is a Kernel in Linux?
   - In simple terms, the kernel is the core or brain of the Linux operating system. It acts as a bridge between the hardware and the software, managing how the computer's resources are used.
   - The kernel in Linux is the core component of the operating system. It’s the part of the system that directly communicates with your computer’s hardware and manages all the resources such as the CPU, memory, storage, and devices. It acts as a bridge between the hardware and the software (applications)Without the kernel, your operating system wouldn’t be able to function — applications wouldn't know how to use the hardware.

        sudo usermod -g g1 p1

        bash

        id p1

8. You suspect that a particular process is consuming excessive CPU resources on your Linux server. How
would you identify and terminate this process?
- **1. Identify the Process Using High CPU**

 You can use tools like top, htop, or ps to check CPU usage.

- Option 1: Using top
       bash

        top
- This shows a real-time list of running processes sorted by CPU usage.

- Look at the %CPU column to identify the culprit.

- Note the PID (Process ID) of the high CPU-consuming process.

Option 2: Using ps

    bash

    ps -eo pid,ppid,cmd,%cpu,%mem --sort=-%cpu | head
- This shows the top CPU-consuming processes in a one-time snapshot.
Option 3: Using htop (if installed)
   bash

    htop
- Interactive and user-friendly.

- You can sort by CPU usage (usually by pressing F6 and selecting %CPU).

- Use arrow keys to highlight the process and F9 to kill.

**2. Investigate the Process (Optional but Recommended)**

Before terminating it, check what the process is:

     bash

      ps -p <PID> -o pid,ppid,cmd,%cpu,%mem

  **3. Terminate the Process**
If you're sure it needs to be stopped:

Gracefully terminate:

    bash
    kill <PID>

Force terminate (if it doesn't stop):

    bash
    kill -9 <PID>

**4. Monitor Again**

Rerun top or ps to confirm the process has stopped and CPU usage has normalized.
